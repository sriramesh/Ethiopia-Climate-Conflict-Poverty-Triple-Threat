```python
df = pd.read_csv('Data/ETH-District-data/Ethiopia_district-data.csv', header=0, skiprows=1)
df = df[['#date','#adm1+name', 'Unnamed: 3', 'Unnamed: 4', '#affected+infected+confirmed', '#affected+infected+dead']]
df = df.rename(index=str, columns={'#date': 'Date',
                                   'Unnamed: 3': 'MPI',
                                   'Unnamed: 4': 'Population',
                                   '#adm1+name': 'Adm1',
                                   '#affected+infected+confirmed': 'COVID-19 Confirmed Cases',
                                   '#affected+infected+dead': 'COVID-19 Reported Deaths'})

df["Date"] = pd.to_datetime(df["Date"])
df = df[(df['Date'] > '2020-10-15 00:00:00')]
df = df[['Adm1', 'MPI', 'Population', 'COVID-19 Confirmed Cases', 'COVID-19 Reported Deaths']]
df['MPI'] = df['MPI'].astype(float)
df['Population'] = df['Population'].astype(int)
df['COVID-19 Confirmed Cases'] = df['COVID-19 Confirmed Cases'].astype(int)
df['COVID-19 Reported Deaths'] = df['COVID-19 Reported Deaths'].astype(int)
df["COVID-19 Cases per 100,000"] = ( df['COVID-19 Confirmed Cases']/df['Population'] ) * 100000
df["COVID-19 Deaths per 100,000"] = ( df['COVID-19 Reported Deaths']/df['Population'] ) * 100000

df = df.replace(to_replace =["SNNPR", 'Addis Ababa', 'Harari', 'Benshangul Gumuz', 'Gambela'],
           value =["Southern Nations, Nationalities and Peoples", 'Addis Abeba', 'Harari People',
                  'Benshangul-Gumaz', 'Gambela Peoples'])

df
```

```python
df.to_csv('Data/ETH-District-data/Ethiopia_district-data-cleaned.csv', index = False, header=True)
```

```python
# join the geodataframe with the cleaned up csv dataframe

merged = eth.set_index('NAME_1').join(df.set_index('Adm1'))

# set a variable that will call whatever column we want to visualise on the map
variable = 'MPI'
colorscale = 'RdYlGn'

# set the range for the choropleth
vmin, vmax = min(merged[variable]), max(merged[variable])

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(10, 6))

# create map
merged.plot(column=variable, cmap=colorscale, linewidth=0.8, ax=ax, edgecolor='0.8')
# plt.title('Ethiopia '+variable)
plt.title('Ethiopia Multidimensional Poverty Index (higher values = higher poverty), 2020')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

sm = plt.cm.ScalarMappable(cmap=colorscale, norm=plt.Normalize(vmin=vmin, vmax=vmax))
cbar = fig.colorbar(sm)

plt.show()
```
